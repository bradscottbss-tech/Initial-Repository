**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
df_comp = pd.read_sql("SELECT CompTotal FROM main", conn)
df_comp['CompTotal'].hist(bins=20)
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
df_age = pd.read_sql("SELECT Age FROM main", conn)
df_age['Age'].plot.box()
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
df_scatter = pd.read_sql("SELECT Age, WorkExp FROM main", conn)
df_scatter.plot(x='Age', y='WorkExp', kind='scatter')
plt.xticks(rotation=90)
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
df_bubble = pd.read_sql("SELECT TimeSearching, Frustration, Age FROM main", conn)

plt.scatter(df_bubble['TimeSearching'], 
            df_bubble['Frustration'], 
            s=df_bubble['Age']*10,  # Multiply Age to make bubbles visible
            alpha=0.5)      # Transparency for overlapping points

plt.xlabel('Time Searching')
plt.ylabel('Frustration')
plt.title('Bubble Plot (Age as size)')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
df_wantwork = pd.read_sql("SELECT DatabaseWantToWorkWith FROM main", conn)

# 1. Get counts, top 5, and handle potential missing values
top_databases = df_wantwork['DatabaseWantToWorkWith'].value_counts().head(5)

# 2. Plotting
plt.figure(figsize=(8, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
df_stacked = pd.read_sql("SELECT TimeSearching, TimeAnswering, Age FROM main", conn)
# 1. Filter for the specific age group
filtered_df = df_stacked[df_stacked['Age'] == "25-34 years old"]

# 2. Calculate medians for the target columns
median_searching = filtered_df['TimeSearching'].median()
median_answering = filtered_df['TimeAnswering'].median()

# 3. Create the stacked bar chart
label = "25-34 years old"
plt.bar(label, median_searching, label='Time Searching')
plt.bar(label, median_answering, bottom=median_searching, label='Time Answering')

# Formatting
plt.ylabel('Median Time')
plt.title('Median Time Spent Searching vs Answering (25-34 years old)')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
df_line = pd.read_sql("SELECT CompTotal, Age FROM main", conn)
df_line.head()

# 1. Filter data for ages 45 to 60
df_filtered = df_line[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)]

# 2. Group by 'Age', calculate median 'CompTotal', and plot
df_filtered.groupby('Age')['CompTotal'].median().plot(kind='line')

# 3. Add descriptive labels
plt.title('Median CompTotal for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median CompTotal')
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
df_bar = pd.read_sql("SELECT MainBranch FROM main", conn)
df_bar.head()

# 1. Count occurrences of each category in 'MainBranch'
branch_counts = df_bar['MainBranch'].value_counts()

# 2. Create the horizontal bar chart
branch_counts.plot(kind='barh', color='skyblue')

# 3. Add titles and labels for clarity
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('Branch Type')

# 4. Show the plot
plt.show()
```
