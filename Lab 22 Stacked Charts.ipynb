### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
df_plot1 = df[['JobSatPoints_6', 'JobSatPoints_7', 'Age']].dropna()
#print(df_plot1.head())
#print(len(df))
#print(len(df_plot1))

#df_plot1
#group by 'Age' category
#find mean of 'JobSatPoints_6' by Age group
#find mean of 'JobSatPoints_7' by Age group
#Age = x-axis
#plot

# 1. Group by 'Age' and calculate the mean for both columns
df_grouped = df_plot1.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].mean()

# 2. Plot the results
df_grouped.plot(kind='bar', figsize=(10, 6))

plt.title('Mean Job Satisfaction Points by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean Points')
plt.legend(["JobSatPoints_6", "JobSatPoints_7"])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here

df_plot2 = df[['JobSatPoints_6', 'JobSatPoints_7', 'Employment']].dropna()
#print(df_plot2.head())
#print(len(df))
#print(len(df_plot2))

df_plot2['Employment_Simp'] = df_plot2['Employment'].str.split(',').str[0].str.strip()
#print(df_plot2.head())

# 1. Group by 'X' and calculate the mean for both columns
df_grouped = df_plot2.groupby('Employment_Simp')[['JobSatPoints_6', 'JobSatPoints_7']].mean()

# 2. Plot the results
df_grouped.plot(kind='bar', figsize=(10, 6))

plt.title('Mean Job Satisfaction Points by Employment Group')
plt.xlabel('Employment Group')
plt.ylabel('Mean Points')
plt.legend(["JobSatPoints_6", "JobSatPoints_7"])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here

df_plot3 = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Age']].dropna()
#print(df_plot3.head())
#print(len(df))
#print(len(df_plot3))

# 1. Group by 'X' and calculate the mean for both columns
df_grouped = df_plot3.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].mean()
df_grouped['ConvertedCompYearly'] = df_grouped['ConvertedCompYearly']/1000

# 2. Plot the results
df_grouped.plot(kind='bar', figsize=(10, 6))

plt.title('Mean Comp and Job Satisfaction Points by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean Points')
plt.legend(["ConvertedCompYearly", "JobSatPoints_6"])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here

df_plot4 = df[['CompTotal', 'JobSat', 'Age']].dropna()
#print(df_plot4.head())
#print(len(df))
#print(len(df_plot4))

# 1. Group by 'X' and calculate the mean for both columns
df_grouped = df_plot4.groupby('Age')[['CompTotal', 'JobSat']].median()
df_grouped['CompTotal'] = df_grouped['CompTotal']/1000

# 2. Plot the results
df_grouped.plot(kind='bar', figsize=(10, 6))

plt.title('Mean Comp and Job Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean Points')
plt.legend(["Comp Total", "Job Sat"])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here

df_plot5 = df[['DatabaseAdmired', 'Age']].dropna()
#print(df_plot5.head())
#print(len(df))
#print(len(df_plot5))

df_5exploded = df_plot5.assign(
    DatabaseAdmired=df_plot5['DatabaseAdmired'].str.split(';')
).explode('DatabaseAdmired')
#print(df_5exploded.head())
#print(len(df_5exploded))

#df_5exploded
#group by 'Age'
#sum each unique 'DatabaseAdmired' for each 'Age' group
#plot count of 'DatabaseAdmired' for each 'Age' group

#Group by 'Age' and count unique DatabaseAdmired occurrences
# size() counts the occurrences including NaN if any, count() excludes NaN
age_db_counts = df_5exploded.groupby(['Age', 'DatabaseAdmired']).size().reset_index(name='Count')

#print(age_db_counts)

#Plot count of 'DatabaseAdmired' for each 'Age' group
# Pivot for better plotting format
plot_df = age_db_counts.pivot(index='DatabaseAdmired', columns='Age', values='Count').fillna(0)

plot_df.plot(kind='bar', figsize=(10, 6))
plt.title('Database Admired by Age Group')
plt.ylabel('Count')
plt.xlabel('Database')
plt.xticks(rotation=45)
plt.legend(title='Age')
plt.tight_layout()
plt.show()
```
