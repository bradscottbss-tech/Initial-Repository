### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

df_plot1 = df[['DatabaseWantToWorkWith']]
#print(df_plot1.head())

#explode DatabaseWantToWorkWith
#sum each DatabaseWantToWorkWith
#find top 5 DatabaseWantToWorkWith with highest sum
#plot top 5 DatabaseWantToWorkWith on pie chart

# 1. Explode the column (assuming data is comma-separated string)
# If already lists, omit .str.split(',')
df_exploded = df_plot1.assign(
    DatabaseWantToWorkWith=df_plot1['DatabaseWantToWorkWith'].str.split(',')
).explode('DatabaseWantToWorkWith')

# Clean up whitespace
df_exploded['DatabaseWantToWorkWith'] = df_exploded['DatabaseWantToWorkWith'].str.strip()

# 2. Count occurrences (sum/count each database)
db_counts = df_exploded['DatabaseWantToWorkWith'].value_counts()

# 3. Find top 5
top5_dbs = db_counts.nlargest(5)

# 4. Plot top 5 on pie chart
plt.figure(figsize=(10, 6))
top5_dbs.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases to Work With')
plt.ylabel('') # Hides y-label
plt.show()
```
