#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
## Write your code here
df.info()

total_missing_values = df.isnull().sum().sum()
print(f"Total missing values in the DataFrame: {total_missing_values}")

missing_counts = df.isnull().sum()
print("Missing values per column:")
print(missing_counts)

print(f"Percent Missing Values: {(total_missing_values/(len(df)*len(df.columns))*100)}%")

missing_employ_values = df['Employment'].isnull().sum()
print(f"Number of missing values in the 'Employment' column: {missing_employ_values}")

missing_jobsat_values = df['JobSat'].isnull().sum()
print(f"Number of missing values in the 'JobSat' column: {missing_jobsat_values}")

missing_remote_values = df['RemoteWork'].isnull().sum()
print(f"Number of missing values in the 'RemoteWork' column: {missing_remote_values}")
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
## Write your code here
import numpy as np

missing_yrscodepro_values = df['YearsCodePro'].isnull().sum()
print(f"Number of missing values in the 'YearsCodePro' column: {missing_yrscodepro_values}")
#print(f"Percent Missing YearsCodePro Values: {(missing_yrscodepro_values/(len(df))*100)}%")

missing_jobsat_values = df['JobSat'].isnull().sum()
print(f"Number of missing values in the 'JobSat' column: {missing_jobsat_values}")
#print(f"Percent Missing JobSat Values: {(missing_jobsat_values/(len(df))*100)}%")

count_yearscodepro = df['YearsCodePro'].nunique()
#print(count_yearscodepro)
value_counts_series_yearscodepro = df['YearsCodePro'].value_counts()
#print(value_counts_series_yearscodepro)

count_jobsat = df['JobSat'].nunique()
print(count_jobsat)
value_counts_series_jobsat = df['JobSat'].value_counts()
print(value_counts_series_jobsat)
##

#Change text to specific numbers
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year': 0, 'More than 50 years': 51})

#Ensure data is numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'])

# Define the bins (edges of the ranges) and labels
bins = [0, 5, 10, 15, 20, 25, 30, 50, np.inf]
labels = ['0-5', '5-10', '10-15', '15-20', '20-25', '25-30', '30-50', '50+']

#Create Ranges (0-5, 5-10, etc.)
# Using 0-5, 5-10, 10-20, 20-50, 50+
bins = [-1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 100]
labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '50+']
df['ExperienceRange'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels)

#Calculate Median JobSat for each range
median_jobsat = df.groupby('ExperienceRange')['JobSat'].median()

print(median_jobsat)
##
# Get the value counts
plt.figure(figsize=(10, 6))
median_jobsat.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Median Job Satisfaction by Years of Professional Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Median Job Satisfaction')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

##
# Get the value counts
series_jobsat = df['JobSat'].value_counts()
# Plot using seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=series_jobsat.index, y=series_jobsat.values)
plt.title('Job Satisfaction Distribution')
plt.ylabel('Count')
plt.xlabel('Job Satisfaction Category')
plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
## Write your code here
missing_remote_values = df['RemoteWork'].isnull().sum()
print(f"Number of missing values in the 'RemoteWork' column: {missing_remote_values}")
print(f"Percent Missing RemoteWork Values: {(missing_remote_values/(len(df))*100)}%")

df['RemoteWork'] = df['RemoteWork'].fillna('Unknown')

count_remote = df['RemoteWork'].nunique()
print(count_remote)
value_counts_series_remote = df['RemoteWork'].value_counts()
print(value_counts_series_remote)

# Plot using seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=value_counts_series_remote.index, y=value_counts_series_remote.values)
plt.title('Remote Work Distribution')
plt.ylabel('Count')
plt.xlabel('Remote Work Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
## Write your code here
#filter dataframe by 'Country'
#identify top 20 'Country' by count
#find most popular 'LanguageHaveWorkedWith' per 'Country'
#print top 20 'Country' and show most popular 'LanguageHaveWorkedWith'

# 1. Identify top 20 'Country' by count
top_20_countries = df['Country'].value_counts().nlargest(20).index

# 2. Filter dataframe for top 20 countries
df_filtered = df[df['Country'].isin(top_20_countries)]

# 3. Handle 'LanguageHaveWorkedWith' (Assuming comma-separated strings)
# Split languages and explode to individual rows
df_lang = df_filtered.assign(
    LanguageHaveWorkedWith=df_filtered['LanguageHaveWorkedWith'].str.split(';')
).explode('LanguageHaveWorkedWith')

# 4. Find most popular 'LanguageHaveWorkedWith' per 'Country'
top_lang_per_country = df_lang.groupby(['Country', 'LanguageHaveWorkedWith']).size().reset_index(name='Count')
top_lang_per_country = top_lang_per_country.sort_values(['Country', 'Count'], ascending=[True, False])
most_popular_lang = top_lang_per_country.groupby('Country').first().reset_index()

# 5. Print result
print(most_popular_lang[['Country', 'LanguageHaveWorkedWith']])
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
## Write your code here
# Factorize string data (assigns a numerical ID to each category)
df['EdLevel_Code'] = pd.factorize(df['EdLevel'])[0]
df['Employment_Code'] = pd.factorize(df['Employment'])[0]

# Calculate Correlation
correlation = df['EdLevel_Code'].corr(df['Employment_Code'])
print(f"Correlation: {correlation}")

plt.scatter(df['EdLevel_Code'], df['Employment_Code'])
plt.xlabel('EdLevel_Code')
plt.ylabel('Employment_Code')
plt.show()
```
