```python
!pip install pandas
!pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)
df.head()
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here

# Select the columns to create a new DataFrame
df_plot1 = df[['Age', 'JobSatPoints_6']]

# Optional: Plot the data (e.g., a scatter plot)
df_plot1.plot(x='Age', y='JobSatPoints_6', kind='scatter')
plt.xticks(rotation=90)
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here

# Select the columns to create a new DataFrame
df_plot2 = df[['ConvertedCompYearly', 'JobSatPoints_6']]

# Optional: Plot the data (e.g., a scatter plot)
df_plot2.plot(y='ConvertedCompYearly', x='JobSatPoints_6', kind='scatter')
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
!pip install seaborn
import seaborn as sns

# 1. Group by 'Age' and 2. find the mean of 'JobSatPoints_6'
df_plot3 = df[['Age', 'JobSatPoints_6']]
df_grouped = df_plot3.groupby('Age')['JobSatPoints_6'].mean().reset_index()

# 3. Plot regression line
sns.regplot(data=df_grouped, x='Age', y='JobSatPoints_6')
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here

# 1. Select columns and drop NaN values
df_plot4 = df[['Age', 'YearsCodePro']].dropna()

# 2. Plot as a scatter plot
df_plot4.plot(x='Age', y='YearsCodePro', kind='scatter')

# 3. Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# 4. Display the plot
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here

# 1. Select columns and drop NaN values
df_plot5 = df[['ConvertedCompYearly', 'JobSatPoints_6']].dropna()

# 2. Plot as a scatter plot
df_plot5.plot(x='JobSatPoints_6', y='ConvertedCompYearly', kind='scatter')

# 3. Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# 4. Display the plot
plt.show()

```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here

df_plot6 = df[['LanguageHaveWorkedWith', 'JobSat']].dropna()
#df_plot6.head()
#count string LanguageHaveWorkedWith which includes multiple per cell
#plot JobSat for top 10 language

# 2. Split and Explode: Turn "Python;Java" into two rows
# Assuming the separator is ';'
df_exploded = df_plot6.assign(
    LanguageHaveWorkedWith=df_plot6['LanguageHaveWorkedWith'].str.split(';')
).explode('LanguageHaveWorkedWith')

# 3. Get top 10 languages
top_10_languages = df_exploded['LanguageHaveWorkedWith'].value_counts().head(10).index

# 4. Filter original exploded data for only top 10 languages
df_top10 = df_exploded[df_exploded['LanguageHaveWorkedWith'].isin(top_10_languages)]

# 5. Plotting
plt.figure(figsize=(12, 8))
sns.countplot(
    data=df_top10, 
    y='LanguageHaveWorkedWith', 
    hue='JobSat', 
    order=top_10_languages
)
plt.title('Job Satisfaction by Top 10 Languages')
plt.xlabel('Count')
plt.ylabel('Programming Language')
plt.legend(title='Job Satisfaction', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here

df_plot7 = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].dropna()
df_plot7.head()

# 1. Create a scatter plot with hue for categorization
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_plot7,
    x='JobSatPoints_6',
    y='ConvertedCompYearly',
    hue='Employment',
    style='Employment',  # Optional: adds different marker shapes
    alpha=0.6,            # Makes points transparent to handle overplotting
    legend=False
)

# 2. Add titles and labels
plt.title('Yearly Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Job Satisfaction (JobSatPoints_6)')
plt.ylabel('Yearly Compensation (USD)')
plt.xticks(rotation=45)
#plt.tight_layout()

# 3. Show the plot
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here

df_plot8 = df[['YearsCodePro', 'Age', 'Country']].dropna()
#df_plot8.head(10)

df_plot8['YearsCodePro'] = pd.to_numeric(df_plot8['YearsCodePro'], errors='coerce')
grouped_data = df_plot8.groupby('Age')['YearsCodePro'].mean()

grouped_data.plot(
    kind='bar', 
    figsize=(10, 6), 
    title='Average YearsCodePro by Age',
    ylabel='Average YearsCodePro',
    xlabel='Age'
)
plt.show()

###

grouped_data = df_plot8.groupby('Age')['Country'].nunique().reset_index()
grouped_data.columns = ['Age', 'Country']
print(grouped_data)

plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Age', y='Country', palette='viridis')
plt.title('Number of Distinct Countries per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Distinct Country Count')
plt.xticks(rotation=45)
plt.show()
```
